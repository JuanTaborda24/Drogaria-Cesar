>### üß† <b>INTRODU√á√ÉO: O que √© Regress√£o Linear?</b>  

A regress√£o linear √© uma t√©cnica estat√≠stica usada para <b>modelar a rela√ß√£o entre uma vari√°vel dependente (alvo)</b> e <b>uma ou mais vari√°veis independentes (preditoras)</b>. Ela √© amplamente usada para <b>previs√£o de valores</b> e <b>an√°lise de impacto de vari√°veis</b>.</p>

>### üìå <b>Por que se chama _Regress√£o Linear_?</b>

- **Regress√£o** significa que estamos tentando **prever ou estimar** um valor com base em outros dados.  
  Por exemplo: prever o **pre√ßo de uma casa** com base na **√°rea**, **dist√¢ncia da praia** e **dist√¢ncia da farm√°cia**.

- **Linear** vem do fato de que a rela√ß√£o entre as vari√°veis √© representada por uma **linha reta**.  
  Ou seja, o modelo tenta ajustar uma **equa√ß√£o do tipo**:

  <img src='https://i.postimg.cc/XY9PvyNz/formula.png'>

Onde:

- <i>Valor</i> √© a vari√°vel dependente (o que queremos prever).

- <i>Area, Dist_Praia, Dist_Farmacia</i> s√£o as vari√°veis independentes.

- ùëé, ùëè, ùëê  s√£o os coeficientes que indicam o impacto de cada vari√°vel.

- ùëë  √© o intercepto (valor base quando todas as vari√°veis s√£o zero).

> √â chamada de **regress√£o linear** porque usamos uma **equa√ß√£o de linha reta** para **estimar ou prever** valores com base em outras vari√°veis.

> ### üìå <b>CONCEITOS IMPORTANTES</b>  

>üîπ <b>Intercepto</b>  

<b>- Statsmodels</b>: O intercepto precisa ser adicionado manualmente com <i>sm.add_constant(X)</i>.   
Aparece como <i>const</i> no resumo estat√≠stico.  

<b>- Scikit-Learn</b>: O intercepto √© inclu√≠do automaticamente.  
Pode ser acessado com <i>model.intercept_</i>.

>üîπ <b>Coeficientes</b>      

<b>- Statsmodels</b>: Os coeficientes aparecem no <i>summary()</i> com seus respectivos p-valores, indicando a signific√¢ncia estat√≠stica de cada vari√°vel.  

<b>- Scikit-Learn</b>: Os coeficientes s√£o acessados com <i>model.coef_</i>, mas n√£o incluem p-valores ou testes estat√≠sticos.

>üîπ <b>M√©tricas estat√≠sticas comparadas</b>

| M√©trica         | `statsmodels`                          | `scikit-learn`                     |
|-----------------|----------------------------------------|------------------------------------|
| R¬≤ (score)      | Inclu√≠do no `summary()`                | `model.score()`                    |
| p-valores       | Sim, para cada vari√°vel                | N√£o dispon√≠vel                     |
| F-statistic     | Sim, avalia o modelo como um todo      | N√£o dispon√≠vel                     |
| Intercepto      | `const` no `summary()`                 | `model.intercept_`                |


>### ‚öñÔ∏è<b>Compara√ß√£o entre `statsmodels` e `scikit-learn`</b>  
>  

| Aspecto                     | `statsmodels`                          | `scikit-learn`                        |
|----------------------------|----------------------------------------|--------------------------------------|
| üì¶ Foco principal           | Estat√≠stica cl√°ssica                   | Machine Learning                     |
| üîç Sa√≠da de resultados      | Resumo estat√≠stico completo            | Apenas coeficientes e previs√µes      |
| üß† Ideal para               | Interpreta√ß√£o e an√°lise profunda       | Previs√£o e integra√ß√£o com ML         |
| üìà Intercepto               | Requer `add_constant()`                | Adicionado automaticamente           |
| üìä M√©tricas estat√≠sticas    | Detalhadas (p-valores, F-stat, etc.)   | Limitadas (R¬≤)                       |

```python
# [Regress√£o Linear 1: c√≥digo com Statsmodels]
# MODELO STATSMODELS
# Selecionar vari√°veis
X = data[['Area', 'Dist_Praia', 'Dist_Farmacia']]
y = data['Valor']


# Adicionar constante (intercepto)
X = sm.add_constant(X)


# Criar e ajustar o modelo
model = sm.OLS(y, X).fit()


# Exibir resumo estat√≠stico
print(model.summary())


```
